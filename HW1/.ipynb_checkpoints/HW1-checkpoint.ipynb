```python
import os, glob
import numpy as np
import pywt
import matplotlib.pyplot as plt
from matplotlib.image import imread
from bs4 import BeautifulSoup
from sklearn.decomposition import NMF
```

```python
ANN_FOLDER = '/Users/alexis/DM/HW1/stanford-dogs-dataset/annotations/Annotation'
IMG_FOLDER = '/Users/alexis/DM/HW1/stanford-dogs-dataset/images/Images'
```

```python
def read_data():
    for ann_file in glob.glob(f'{ANN_FOLDER}/*/*'):
        filepath = ann_file.split('/')[-2:]
        filepath = '/'.join(filepath) + '.jpg'
        img_file = os.path.join(IMG_FOLDER, filepath)

        image = imread(img_file)
        image = np.float64(image)

        ann_f = open(ann_file, "r")
        ann_txt = ann_f.read()
        ann_soup = BeautifulSoup(ann_txt, 'html.parser')
        object_soup = ann_soup.select('annotation object')

        result = dict()
        result['img'] = image
        result['id'] = ann_soup.select('annotation filename')[0].text
        result['name'] = ann_soup.select('annotation object name')[0].text
        result['xmin'] = ann_soup.select('annotation object bndbox xmin')[0].text
        result['ymin'] = ann_soup.select('annotation object bndbox ymin')[0].text
        result['xmax'] = ann_soup.select('annotation object bndbox xmax')[0].text
        result['ymax'] = ann_soup.select('annotation object bndbox ymax')[0].text

        yield result
```

```python
def run_dwt(image):
    new_image = []
    for c in range(image.shape[2]):
        coeffs = pywt.dwt2(image[:,:,c], 'haar')
        cA, (cH, cV, cD) = coeffs
        new_image.append(cA)

    new_image = np.array(new_image).transpose(1, 2, 0)
    # matplotlib.pyplot.imshow(new_image)
    # matplotlib.pyplot.show()
    return new_image

def run_svd(image):
    num_components = 10
    new_image = []
    for c in range(image.shape[2]):
        u, s, vh = np.linalg.svd(image[:,:,c], full_matrices=True)
        reconst_img = np.matrix(u[:, :num_components]) * np.diag(s[:num_components]) * np.matrix(vh[:num_components, :])
        new_image.append(reconst_img)

    new_image = np.array(new_image).transpose(1, 2, 0)
    # matplotlib.pyplot.imshow(new_image)
    # matplotlib.pyplot.show()
    return new_image

def run_nmf(image):
    num_components = 10
    new_image = []
    for c in range(image.shape[2]):
        model = NMF(num_components, init='random', random_state=0)
        W = model.fit_transform(image[:,:,c])
        H = model.components_
        reconst_img = np.matmul(W,H)
        new_image.append(reconst_img)

    new_image = np.array(new_image).transpose(1, 2, 0)
    # matplotlib.pyplot.imshow(new_image)
    # matplotlib.pyplot.show()
    return new_image
```

```python
plt.rcParams['figure.dpi'] = 170
```

```python
generator = read_data()
```

```python
plt.imshow(img.astype('uint8'))
plt.show()
```

```python
a = run_dwt(img)
plt.imshow(a.astype('uint8'))
plt.show()
```

```python
b = run_svd(img)
plt.imshow(b.astype('uint8'))
plt.show()
```

```python
c = run_nmf(img)
plt.imshow(c.astype('uint8'))
plt.show()
```
